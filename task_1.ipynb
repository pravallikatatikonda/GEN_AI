Task 1: Conceptual Understanding
1. Difference between "Love" and "love" in NLP

In Natural Language Processing (NLP), "Love" and "love" are treated as two different tokens because NLP systems are case-sensitive by default. This increases the vocabulary size unnecessarily and can reduce model efficiency. To avoid this issue, text normalization techniques such as converting all text to lowercase are applied so that both words are treated the same.

2. What happens if stopwords are not removed?

If stopwords are not removed, the dataset will contain many common words such as "is", "the", and "and" that do not carry significant meaning. This increases noise in the data, makes processing slower, and may reduce the accuracy of machine learning models because important words get less importance.

3. Two real-world scenarios where removing stopwords can be harmful
Sentiment Analysis:
Removing stopwords like "not" can completely change the meaning of a sentence.
Example: "This is not good" → "This good" (wrong meaning)
Question Answering Systems:
Removing words like "what", "where", and "how" removes the intent of the question, making it difficult for the system to understand the query.
4. Difference between Stemming and Lemmatization
Stemming:
It reduces words to their root form by removing suffixes. The result may not always be a valid word.
Example: "running" → "run", "studies" → "studi"
Lemmatization:
It reduces words to their base or dictionary form using vocabulary and grammar rules. The result is always meaningful.

```python
# Task 2: preprocess_text()
import re

def preprocess_text(text):
    if not text:
        return [], ""

    # Remove URLs
    text = re.sub(r"http\S+", "", text)

    # Remove numbers
    text = re.sub(r"\d+", "", text)

    # Convert to lowercase
    text = text.lower()

    # Handle repeated characters (soooo → soo)
    text = re.sub(r'(.)\1+', r'\1\1', text)

    # Remove special characters
    text = re.sub(r'[^a-z\s]', '', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # Tokenization
    tokens = text.split()

    # Remove short tokens (≤2) except 'no', 'not'
    tokens = [w for w in tokens if len(w) > 2 or w in ['no', 'not']]

    cleaned_sentence = " ".join(tokens)

    return tokens, cleaned_sentence
```

```python
#Task 3: Stress Testing
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

for text in test_sentences:
    tokens, clean = preprocess_text(text)

    print("\nOriginal Text:", text)
    print("Cleaned Tokens:", tokens)
    print("Cleaned Sentence:", clean)
```

```python
#Task 4: Token Analytics
def token_analysis(tokens):
    total = len(tokens)
    unique = len(set(tokens))
    avg_len = sum(len(w) for w in tokens) / total if total > 0 else 0

    return total, unique, avg_len

for text in test_sentences:
    tokens, _ = preprocess_text(text)
    total, unique, avg_len = token_analysis(tokens)

    print("\nSentence:", text)
    print("Total Tokens:", total)
    print("Unique Tokens:", unique)
    print("Average Token Length:", avg_len)
```

```python
#Task 5: Frequency Analysis
from collections import Counter

all_tokens = []

for text in test_sentences:
    tokens, _ = preprocess_text(text)
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("Top 10 frequent words:", counter.most_common(10))
print("Top 5 least frequent words:", counter.most_common()[-5:])
```

```python
#Task 6: Full Pipeline
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```
